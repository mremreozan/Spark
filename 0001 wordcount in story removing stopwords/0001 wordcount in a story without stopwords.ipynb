```python
findspark.init('C:\spark')    # we show path of pyspark to jupyter
```

```python
from pyspark import SparkContext
```

```python
sc = SparkContext()
```

```python
lines = sc.textFile('Aladdin and the Wonderful Lamp.txt')
```

```python
lines.count()
```

```python
lines.take(2)
```

```python
words = lines.flatMap(lambda x : x.split(' '))
```

```python
words.count()
```

```python
words.take(3)
```

```python
stopwords = sc.textFile('stopwords.txt')
```

```python
stopwords = stopwords.collect()  # we transform to list from RDD for using stopwords with filter function
```

```python
filteredWords = words.filter(lambda x: x.lower() not in stopwords)
```

```python
empty = ['', ' ', 'a', 'i']
```

```python
filteredWords = filteredWords.filter(lambda x: x not in empty)
```

```python
filteredWords.count()
```

```python
pairs = filteredWords.map(lambda x: (x,1))
```

```python
pairs.take(2)
```

```python
from operator import add 
```

```python
wordname = pairs.reduceByKey(add)  # add -> lambda x,y: x+y
```

```python
wordname.count()
```

```python
wordname.take(2)
```

```python
output = wordname.sortBy((lambda x : x[1]), False).collect()   # we must use sortBy, not sortByKey
```

```python
output[0:5]
```

```python
for (name, count) in output: print (name, count)
```
